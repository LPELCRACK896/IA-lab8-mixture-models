```python
import pandas as pd
df = pd.read_csv('bank_transactions.csv', nrows=100)
```

```python
# Codificación de variables categóricas
df_encoded = pd.get_dummies(df, columns=['TransactionID'])
df_encoded = pd.get_dummies(df, columns=['CustomerID'])
df_encoded = pd.get_dummies(df, columns=['CustomerDOB'])
df_encoded = pd.get_dummies(df, columns=['CustGender'])
df_encoded = pd.get_dummies(df, columns=['CustLocation'])
df_encoded = pd.get_dummies(df, columns=['CustAccountBalance'])
df_encoded = pd.get_dummies(df, columns=['TransactionDate'])
df_encoded = pd.get_dummies(df, columns=['TransactionTime'])
df_encoded = pd.get_dummies(df, columns=['TransactionAmount (INR)'])
```
